```python
#Exercise 7
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, 'tab:blue')
ax.set(title='y = sin(x)')

fig, ax = plt.subplots()
ax.scatter(x, y2, color='blue')
ax.set(title='y = cos(x)')
plt.show()
```

```python
#Exercise 9
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='y = sin(x)')
ax.legend()

fig, ax = plt.subplots()
ax.scatter(x, y2, color='blue', label='y = cos(x)')
ax.legend()
plt.show()
```
