```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Task 1
x = np.linspace(0, 20, 100)
y = x**2
plt.plot(x, y)
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 20)  # Set x-axis limits
plt.ylim(0, 400)  # Set y-axis limits
plt.show()
plt.savefig('plot.png')
```

```python
#Task 2
df = pd.read_csv('chipo.txt', sep='\t', header=None)
df.to_csv('chipo.csv', header=None)

chipo = pd.read_csv('chipo.csv')
chipo.head()
```

```python
# Remove the dollar sign and convert to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace('$', '')))

print(chipo.head())
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(chipo['item_price'], chipo['quantity'], alpha=0.5, edgecolor='k')
plt.title('Relationship between Item Price and Quantity Ordered')
plt.xlabel('Item Price ($)')
plt.ylabel('Quantity Ordered')
plt.show()
plt.savefig('plot1.png')
```

```python
#Task 3
item_totals = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
print(item_totals.head())
```

```python
top_5_items = item_totals.head(5)

plt.figure(figsize=(12, 8))
top_5_items.plot(kind='bar', color='blue')
plt.title('Top 5 Items by Total Quantity Ordered')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')
plt.show()
plt.savefig('plot2.png')
```
